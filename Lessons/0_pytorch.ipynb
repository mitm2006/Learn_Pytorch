## Tensor Datatypes

```python
float_32_tensor=torch.tensor([3.0,6.0,9.0],
                             dtype=None,
                             device=None,
                             requires_grad=False)
float_32_tensor.dtype
```

```python
float_16_tensor=float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_multi=float_16_tensor*float_32_tensor
float_multi
```
